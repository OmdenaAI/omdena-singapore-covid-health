```
import requests


r=requests.get('https://www.imf.org/en/Topics/imf-and-covid19/Policy-Responses-to-COVID-19')
c=r.content

from bs4 import BeautifulSoup

soup=BeautifulSoup(c)
```

```
print(soup)
```

```
main_content=soup.find('div',attrs={'class': 'rr-intro'})
main_content
```

```
countries_tags=main_content.find_all('h3')
countries=list()
for c in countries_tags:
  countries.append(c.get_text())
countries_tags.clear
```

```
comments_tags=main_content.find_all('p')
comments_tags=comments_tags[3:]
comments=list()
for c in comments_tags:
  comments.append(c.get_text())
comments_tags.clear
```

```
len(countries)
```

```
len(comments)
```

```
context=list()
for idx,c in enumerate(comments):
  if c=="Back to Top":
    context.append(comments[idx+1])
context
```

```
listed_tags=main_content.find_all('ul')
listed_tags=listed_tags[1:]
listed_tags
```

```
len(listed_tags)
```

```
fiscal=list()
monetary=list()
exchange_rate=list()
for i in range(len(countries)):
  if i<135:
    fiscal.append(listed_tags[3*i+0].get_text().strip())
    monetary.append(listed_tags[3*i+1].get_text().strip())
    exchange_rate.append(listed_tags[3*i+2].get_text().strip())
  elif i==135:
    fiscal.append(listed_tags[3*i+0].get_text().strip())
    monetary.append(listed_tags[3*i+1].get_text().strip()+listed_tags[3*i+2].get_text().strip())
    exchange_rate.append(listed_tags[3*i+3].get_text().strip())
  elif i>135:
    fiscal.append(listed_tags[3*i+1].get_text().strip())
    monetary.append(listed_tags[3*i+2].get_text().strip())
    exchange_rate.append(listed_tags[3*i+3].get_text().strip())
```

```
countries[135]
```

```
updated_dates=main_content.find_all('h6')
updated_dates=[i.get_text().replace('Key Policy Responses as of ','') for i in updated_dates]
updated_dates
```

```
len(updated_dates)
```

```
import pandas as pd
```

```
df=pd.DataFrame(list(zip(countries,updated_dates,context,fiscal,monetary,exchange_rate)),columns=['country','updated date','comments','fiscal','monetary and macro-finacial','exchange rate and balance of payments'])
```

```
df
```

```
df.to_csv('COVID-19 country policies.csv',index=False)
```
