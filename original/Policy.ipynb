```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
data=pd.read_excel('/content/Austria.xlsx',sheet_name='sector')
```

```
plt.style.use('seaborn')
date=data['Unnamed: 0']
construction=data['Construction']
manufacture=data['Manufacture ']
trade=data['Trade']
transport=data['Transportation and storage']
accomod=data['Accomodation and gastronomy']
health=data['Health and Social care']
plt.plot_date(date,   construction,   linestyle='solid' ,color='pink', label='Construction')
plt.plot_date(date,   trade,   linestyle='solid' ,color='blue', label='Trade')
plt.plot_date(date,   transport,   linestyle='solid' ,color='brown', label='Transport')
plt.plot_date(date,   accomod,   linestyle='solid' ,color='yellow', label='Accomodation')
plt.plot_date(date,   manufacture,   linestyle='solid' ,color='grey', label='manufacture')
plt.plot_date(date,   health,   linestyle='solid' ,color='purple', label='Health and Social care')
plt.title('Austria')
plt.xlabel('date')
plt.ylabel('unemployed persons')
plt.tight_layout()
plt.legend()
plt.show()
```

```
data2=pd.read_excel('/content/Austria.xlsx',sheet_name='special persons')
```

```
data2.head()
```

```
plt.style.use('seaborn')
date=data2['Unnamed: 0']
nationals=data2['Nationals']
foreigners=data2['Foreigners']
plt.plot_date(date,   nationals,   linestyle='solid' ,color='grey', label='nationals')
plt.plot_date(date,  foreigners,   linestyle='solid' ,color='purple', label='foreigners')
plt.title('Austria')
plt.xlabel('date')
plt.ylabel('unemployed persons')
plt.legend()
plt.show()
```

```
plt.style.use('seaborn')
date=data2['Unnamed: 0']
above25=data2['25 to 49']
below25=data2[' 25 and below']
above50=data2['50 and above ']
plt.plot_date(date,  above25,   linestyle='solid' ,color='purple', label='25 to 49')
plt.plot_date(date,  below25,   linestyle='solid' ,color='black', label='25 and below')
plt.plot_date(date,  above50,   linestyle='solid' ,color='grey', label='50 and above')
plt.title('Austria')
plt.xlabel('date')
plt.ylabel('unemployed persons')
plt.legend()
plt.show()
```

```
plt.style.use('seaborn')
date=data2['Unnamed: 0']
disabled=data2['Disabled people']
nohealth_conditions=data2['People without health restrictions']
health_conditions=data2['People with other health limitations']
plt.plot_date(date,  disabled,   linestyle='solid' ,color='purple', label='Disabled people')
plt.plot_date(date,  nohealth_conditions,   linestyle='solid' ,color='black', label='People without health restrictions'	)
plt.plot_date(date,  health_conditions,   linestyle='solid' ,color='grey', label='People with other health limitations')
plt.title('Austria')
plt.xlabel('date')
plt.ylabel('unemployed persons')
plt.legend()
plt.show()
```
