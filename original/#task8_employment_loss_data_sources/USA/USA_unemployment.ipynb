```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# USA Unemployment
Labor Force Statistics from the Current Population Survey<br>

Series Id:           LNS14000000<br>
Seasonally Adjusted<br>
Series title:        (Seas) Unemployment Rate<br>
Labor force status:  Unemployment rate<br>
Type of data:        Percent or rate<br>

Labor force status:	Civilian labor force			
Type of data:	Number in thousands<br>

Labor force status:	Employed			
Type of data:	Number in thousands<br>

Labor force status:	Unemployed			
Type of data:	Number in thousands<br>

Age:                 16 years and over<br>


```python
df_unemp =pd.read_csv('USA_general_unemployment_SA.csv')
df_unemp['date']=pd.date_range(start='1/1/2006', periods=len(df_unemp), freq='M')
df_unemp=df_unemp.set_index('date')
df_unemp.head()
```

```python
fig=plt.figure(figsize=(12,6))

plt.plot(df_unemp[df_unemp.index.year>=2014]['unemployment_rate'],label='Unemployment rate USA, percent', lw=2)

plt.xlabel('Date')
plt.ylabel('Percent %')

plt.legend()

plt.show()
```

# Unemployment by economical sector

```python
df_sector=pd.read_csv('USA_unemployment_by_sector_NSA.csv')
df_sector['date']=pd.date_range(start='1/1/2006', periods=len(df_sector), freq='M')
df_sector=df_sector.set_index('date')
df_sector
```

```python
fig=plt.figure(figsize=(12,6))

plt.plot(df_sector[df_sector.index.year>=2015]['Total, 16 years and over (percentage)'],
         label='Unemployment rate USA (%)', lw=2)
plt.plot(df_sector[df_sector.index.year>=2015]['Financial activities (percentage)'],
        label='Unemployment rate USA, Finantial activities (%)', lw=2)
plt.plot(df_sector[df_sector.index.year>=2015]['Manufacturing (percentage)'],
         label='Unemployment rate USA, Manufacturing (%)', lw=2)
plt.plot(df_sector[df_sector.index.year>=2015]['Leisure and hospitality (percentage)'],
         label='Unemployment rate USA, Leisure and hospitality (%)', lw=2)



plt.xlabel('Date')
plt.ylabel('Percent %')

plt.legend()

plt.show()
```
