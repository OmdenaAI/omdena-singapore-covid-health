```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# General unemployment data, Sweden

Unemployed as percentage of labour force. The labour force consists of employed and unemployed persons.

```python
pwd
```

```python
df1 = pd.read_excel('unemployed_seasonally_adjusted_M2013_2020_.xlsx', index_col=0)
df1=df1.transpose()
df1['date']=pd.date_range(start='1/1/2013', periods=len(df1), freq='M')
df1=df1.set_index('date')
print('whole df1 dataset: {}'.format(df1.shape))
df1.head(12)
```

```python
fig=plt.figure(figsize=(12,6))

plt.plot(df1['Unemployment rate, percent'],label='Unemployment rate, percent', lw=2)

plt.xlabel('Date')
plt.ylabel('Percent %')

plt.legend()

plt.show()
```

# Employment and Unemployment by region, Sweden

Unemployed as percentage of labour force. The labour force consists of employed and unemployed persons.<br>
Employed as a percentage of the population.

```python
df2 = pd.read_excel('population_percent_region_labour_status_sex_gender_Q2005_2019_.xlsx', index_col=0)
df2=df2.transpose()
df2['date']=pd.date_range(start='4/1/2005', periods=len(df2), freq='Q')
df2=df2.set_index('date')
print('whole df2 dataset: {}'.format(df2.shape))
df2.tail()
```

```python
fig=plt.figure(figsize=(12,6))

plt.plot(df2['Stockholm unemployed'],label='Unemployment rate Stockholm, percent', lw=2)

plt.xlabel('Date')
plt.ylabel('Percent %')

plt.legend()

plt.show()
```

# Unemployment by age and gender, Sweden

```python
df3 = pd.read_excel('unemployed_percent_sex_age_M2005_2020_.xlsx', index_col=0)
df3=df3.transpose()
df3['date']=pd.date_range(start='4/1/2005', periods=len(df3), freq='M')
df3=df3.set_index('date')
print('whole df2 dataset: {}'.format(df3.shape))
df3.tail()
```

```python
fig=plt.figure(figsize=(12,6))

plt.plot(df3['unemployed 15-24 years total'],label='Unemployment rate under 24 years old, percent', lw=1.5)
plt.plot(df3['unemployed 15-74 years total'],label='Unemployment rate years old, percent', lw=1.5)

plt.xlabel('Date')
plt.ylabel('Percent %')

plt.legend()

plt.show()
```
