```python
import pandas as pd
import numpy as np
```

```python
data_df=pd.read_excel('Share_of_sector_to_total_by_country.xlsx')
data_df
```

```python
aggregate_df = data_df[np.array([data_df.classif1.iloc[i].startswith('Aggregate') for i in range(len(data_df))], dtype=bool)]
#aggregate_df.to_excel('Share_of_sector_to_total_by_country_.xlsx', index=False)
```

```python
detailed_df = data_df[np.array([data_df.classif1.iloc[i].startswith('Detailed') for i in range(len(data_df))], dtype=bool)]
detailed_df
```

```python
#add percentage for each country and each gender
df=pd.DataFrame()
for country in np.unique(detailed_df.ref_area):
    for sex in np.unique(detailed_df.sex):
        df_filter=detailed_df.query('ref_area==@country & sex==@sex')
        total_people=df_filter[df_filter.classif1=='Detailed: Total'].obs_value.values
        #print(country, sex,total_people)
        percent=df_filter.obs_value.values/total_people
        df_filter['percent']=percent
        df=pd.concat([df,df_filter])
```

```python
#add percentage for each country and each gender
df=pd.DataFrame()
for country in np.unique(aggregate_df.ref_area):
    for sex in np.unique(aggregate_df.sex):
        df_filter=aggregate_df.query('ref_area==@country & sex==@sex')
        total_people=np.sum(df_filter.obs_value.values)
        percent=df_filter.obs_value.values/total_people
        df_filter['percent']=percent
        df=pd.concat([df,df_filter])
```

```python
df=df.rename(columns={'classif1':'sector', 'obs_value':'thousands'})
df
```

```python
df.to_excel('detailed_share_of_sector_to_total_by_country_.xlsx', index=False)
```
