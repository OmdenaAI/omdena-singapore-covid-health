```python
from twitterscraper import query_tweets
import pandas as pd
import datetime as dt
```

```python
begin_date = dt.date(2020,1,1)
end_date = dt.date(2020,5,26)
limit=30000
lang = 'english'

```

```python
tweets = query_tweets("#metoo",begindate=begin_date,enddate=end_date,limit=limit,lang=lang)
```

```python
df = pd.DataFrame(t.__dict__ for t in tweets)
```

```python
df.head()
```

```python
df.to_csv('#metoo_twitter.csv')
```
