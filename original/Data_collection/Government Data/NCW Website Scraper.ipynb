### Crime Nature wise Scraping

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import os
import re
```

```python
filepath = "D:/Omdena"
folder_name = "NCW_Data"
nature_folder = "Nature"
state_folder = "State"
os.chdir(filepath + "/" + folder_name)
```

```python
os.listdir()
```

```python
def create_nature_dataframe(year = 2020):
    base_url = "http://164.100.58.238/frmReportNature.aspx?Year=" + str(year)
    response = requests.get(base_url)
    print("response code : ", response)
    
    print("Creating Soup...")
    # creating soup
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # finding base table
    base_table = soup.find('table', class_ = 'NCWDataGridBody')
    
    print("Scraping Data from the Base table...")
    # scraping table entries from the base table
    elements = ""
    rows = []
    for row in base_table.find_all('tr'):
        elements = " "
        for content in row.find_all('td'):
            lines = []
            elements += content.get_text('td') + " "
            lines.append(elements.split(" "))
        rows.append(lines)
    
    print("Reorganizing and cleaning the data...")
    # reorganizing and cleaning the scraped data
    data = dict()
    for element in rows[1:-1]:
        nature = ''
        crime_counts = []
        for e in element[0][2:-1]:
            if e.isnumeric():
                crime_counts.append(int(e))
            else:
                e = re.sub("\ntd", '', e)
                e = re.sub("td\n", '', e)
                nature += e + ' '
        data[nature] = crime_counts
    
    nature_df = pd.DataFrame(data).T
    nature_df.reset_index(inplace = True)
    nature_df.columns = rows[0][0][2:-1]
    
    print("DataFrame created successfully...")
    return nature_df
```

```python
for year in range(2000, 2021):
    df = create_nature_dataframe(year)
    df.to_csv(nature_folder + "/nature_" + str(year) + ".csv", index = False)
    print("Dataframe for year {} saved successfully.".format(year))
    print("\n")
```
