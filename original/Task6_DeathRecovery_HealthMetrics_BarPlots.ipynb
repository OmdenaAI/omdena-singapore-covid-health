```
death_rate = df['death_rate'].tolist()
recovered_rate = df['recovered_rate'].tolist()
haq = df['Health_Access_Quality_Index_2016 (over 100)'].tolist()
prob_3070 = df['Probability_of_Dying_between_30_70_from_chronic_diseases'].tolist()
health_exp = df['Health_Expenditure_per_GDP'].tolist()
physicians = df['Physicians_per_10000_population'].tolist()
beds = df['Critical_Beds_per_10000_Population'].tolist()
pop_over_65 = df['population_over_65_in_%(2018)'].tolist()
population = df['Population_2019'].tolist()
cases_avg = df['cases_avg'].tolist()
recovered_avg = df['recovered_avg'].tolist()
death_avg = df['death_avg'].tolist()

for i in range(0, 10) : 
  death_rate[i] = death_rate[i] * 100
for i in range(0,10) : 
  recovered_rate[i] = recovered_rate[i] * 100
```

```
# Population over 65 in % / recovery rate
# Population over 65 in % / death rate

pop_death_ratio = [0] * 10
pop_recov_ratio = [0] * 10
infected_pop = [0] * 10
dead_pop = [0] * 10
rec_pop = [0] * 10
for i in range(0, 10) : 
  infected_pop[i] = cases_avg[i] / population[i] * 100 # % of people infected
  dead_pop[i] = death_avg[i] / population[i] * 100 # % of people dead due to covid-19
  rec_pop[i] = recovered_avg[i] / population[i] * 100 # % of people recovered from covid-19

infected_65 = [0] * 10
for i in range(0, 10) :
  infected_65[i] = infected_pop[i] / pop_over_65[i] # % of people infected and over 65

for i in range(0, 10) :
  if(dead_pop[i] != 0) : pop_death_ratio[i] = dead_pop[i] / infected_65[i] # % of people dead due to covid / % of people infected & over 65
  if(rec_pop[i] != 0) : pop_recov_ratio[i] = rec_pop[i] / infected_65[i] # % of people recovered from covid / % of people infected & over 65

#print(pop_death_ratio)
#print(pop_recov_ratio)

X = np.arange(10)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(X + 0.00, pop_death_ratio, color = 'red', width = 0.25, label='Population over 65/Death rate')
ax.bar(X + 0.25, pop_recov_ratio, color = 'green', width = 0.25, label='Population over 65/Recovery rate')
plt.xlabel('Countries')
plt.legend(loc="best")
plt.title('Population over 65 death/recovery ratios')
plt.axis('tight')
plt.style.context('seaborn-white')
plt.xticks(ind, countries, rotation=90)
plt.yticks(np.arange(0, 12, 2))
plt.show()
```
