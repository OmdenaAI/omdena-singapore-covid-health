```
# Dataset download link : https://drive.google.com/file/d/1iH3MFApMtwtwjFkyKBhMeXKWwp5UrZTf/view?usp=sharing
link = '/content/drive/My Drive/Task9_MigrationPolicies.csv'
df = pd.read_csv(link, header=0, engine = 'python', skiprows=[2,7])
df.head()
```

```
df.tail()
```

```
countries = ['Australia', 'Denmark', 'India', 'Italy', 'Japan', 'South Korea', 'Sweden', 'United Kingdom', 'United States']
```

```
migrants_per = df['Percentage of Migrants'].tolist()
migrants_no = df['Number of Migrants'].tolist()
population = df['Total Population'].tolist()
migration_policy = df['Migration Policy'].tolist()
```

```
N = 9
ind = np.arange(9) 
width = 0.35  

#migrants_no = np.array(migrants_no)
mig_pol = np.array(migration_policy)
#yes_pol = mig_pol == 1
#no_pol = mig_pol == 0
my_color = np.where(mig_pol==1, 'deepskyblue', 'crimson')
my_label = np.where(mig_pol==1, 'Migrants Support Policy', 'No Policy for Migrants')

fig, ax = plt.subplots(figsize=(24,12))
ax.bar(ind, migrants_no, width, color=my_color)
ax.bar(ind, [0,0,0,0,0,0,0,0,0], width, color='deepskyblue')
#ax.bar(ind[yes_pol], migrants_no[yes_pol], width, color='b', label='yes pol')
#ax.bar(ind[no_pol], migrants_no[no_pol], width, color='r', label='no pol')
plt.ylabel('Number of Migrants', fontsize=20)
plt.xlabel('Countries', fontsize=20)
plt.legend(['No Policy', 'Migrants Support Policy'], loc="best", fontsize=20)
plt.title('Impact of Migration Policies on Migrants during COVID-19', fontsize=24)
plt.axis('tight')
plt.style.context('seaborn-white')
plt.xticks(ind, countries, rotation=45, fontsize=20)
plt.show()
```
