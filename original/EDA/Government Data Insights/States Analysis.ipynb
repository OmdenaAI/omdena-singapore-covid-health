### Instructions for reading the Map:

- The Maps are Interactive, you can zoom in, zoom out on on the whole Map
- The Yellow dots corresponds to their respective states
- Hovering and Clicking over the Yellow dots will give you the count of crimes for the respective state
- The Size of the Circle represents the count of crime occured on the respective state
- The Maps use data only for the First half of the year to give better comparative insights of crime against women among different years.
- The data has count of total crimes recorded against women therefore cannot be specifically said as all the crime counts recorded are domestic violence, although the data does includes the count of domestic violence cases.


### Geographical heatmap for Crime in Different States against Women reported in Year 2020

```python
center_lat = 20.5937
center_long = 78.9629
m = folium.Map(location = [center_lat, center_long], zoom_start = 4.5, height = 700)

for state in df_2020['State_2020'].unique():
    if state.strip() != "Total" and df_2020[df_2020['State_2020'] == state]['Total_2020'].values[0] != 0.0:
        total_2000 = df_2020[df_2020['State_2020'] == state]
        folium.CircleMarker(
            location = [states[states['states'] == state.strip()]['lat'].values[0], states[states['states'] == state.strip()]['long'].values[0]],
            radius = float(df_2020[df_2020['State_2020'] == state]['Total_2020'].values[0])/120,
            popup = "Total Crime : {}".format(df_2020[df_2020['State_2020'] == state]['Total_2020'].values[0]),
            tooltip = state.strip(),
            color = "orange",
            fill_color = "orange",
            legend_name = "Crime Rate"
        ).add_to(m)

folium.FitBounds([(center_lat-14.25, center_long-10), (center_lat+14.25, center_long+10)]).add_to(m)
m
```
