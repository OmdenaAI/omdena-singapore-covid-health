```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_json('cases_monthly_statewise.json')
data.to_csv('cases_monthly_statewise.csv')
```

```python
df = pd.read_csv('cases_monthly_statewise.csv')
```

```python
pd.set_option('display.max_rows', df.shape[0])
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

<h1>1. Analysis of Total Cases each year.</h1>

```python
df_total = df[df['State'] == 'Total']
df_total['Total'] = df_total['Total'].astype(str).astype(int)
df_total.head()
```

```python
plt.figure(figsize=(12,5))
sns.barplot(df_total['Year'], df_total['Total'])
```

<p>As shown here in the graph, 2014 seems to have the highest number of cases after which there is a decrease in cases in 2015.</p>

<h1>2. Analysis of the states in 2020</h1>

```python
df_2020 = df[df['Year'] == 2020]
```

```python
list_i = []
for i in range(df_2020.shape[0]):
    list_i.append(i)
df_2020['index_no'] = list_i
```

```python
df_2020['Total'] = df_2020['Total'].astype('str').astype('int')
df_2020
```

```python
plt.figure(figsize=(15,5))
sns.barplot(df_2020['index_no'], df_2020['Total'])
```

<p>The above graph shows that Uttar Pradesh(28) has the highest number of cses of domestic violence.</p>

<h1>3. Monthwise Analysis in 2020</h1>

```python
months = [df_2020['Jan'].astype('str').astype('int').sum(), df_2020['Feb'].astype('str').astype('int').sum(), df_2020['Mar'].astype('str').astype('int').sum(), df_2020['Apr'].astype('str').astype('int').sum(), df_2020['May'].astype('str').astype('int').sum()]
m = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
plt.figure(figsize=(15,5))
sns.barplot( m, months)
plt.show()
```

<h1>4. Analysis of Uttar Pradesh throughout the years</h1>

```python
df_UP = df[df['State'] == 'Uttar Pradesh']
df_UP['Total'] = df_UP['Total'].astype('str').astype('int')
plt.figure(figsize=(15,5))
sns.barplot(df_UP['Year'], df_UP['Total'])
```

<h1>5. Analysis of all the states throughout the years.</h1>

```python
states_uni = df['State'].unique()
for i in range(len(states_uni)):
    df_state = df[df['State'] == states_uni[i]]
    df_state['Total'] = df_state['Total'].astype('str').astype('int')
    plt.figure(figsize=(15,5))
    plt.title(states_uni[i])
    sns.barplot(df_state['Year'], df_state['Total'])
```
