```python
import pandas as pd
import matplotlib 
import matplotlib.pylab as plt
import numpy as np
```

```python
unemployment = pd.read_excel('#task7_wage_loss_data_source/IndividualDescriptorData/unemployment_rate_monthly.xlsx')
```

```python
unemployment[unemployment['ref_area.label']== 'Sweden'].head(20)
                                                        
```

```python
# Selecting countries

unemployment = unemployment[(unemployment['ref_area.label']=='Australia') | (unemployment['ref_area.label']=='Denmark')
            | (unemployment['ref_area.label']=='Japan')
            | (unemployment['ref_area.label']=='Korea, Republic of')
            | (unemployment['ref_area.label']=='Sweden')
            | (unemployment['ref_area.label']=='United States')]
```

```python
# Selecting age band (Aggregated band, 10 years band, Youth/adults)

unemployment = unemployment[(unemployment['classif1.label']=='Age (Youth, adults): 15+') 
                         | (unemployment['classif1.label']=='Age (Youth, adults): 15-64')
                         | (unemployment['classif1.label']=='Age (Youth, adults): 15-24')                                      
                         | (unemployment['classif1.label']=='Age (Youth, adults): 25+')]
```

```python
# Selecting country, Gender and Age Band
# Country: ref_area.label, Gender: sex.label, Age: classif1.label, Date: time, Unemployment: obs_value

def select_info(country,gender,age):
    
    selected = unemployment[(unemployment['ref_area.label']==country) 
                         & (unemployment['sex.label']== gender)
                         & (unemployment['classif1.label']== age)]

    return selected

```

```python
select_info('Japan','Sex: Total','Age (Youth, adults): 15+')
```

```python
# Plots for countries (total, 15-24, +25)


#ages = ['Age (Youth, adults): 15+', 'Age (Youth, adults): 15-64', 
#        'Age (Youth, adults): 15-24','Age (Youth, adults): 25+']


countries = ['Australia','Denmark','Japan','Korea, Republic of', 'Sweden','United States']

genders = ['Sex: Total','Sex: Male','Sex: Female']


for country in countries:
   
    info1 = select_info(country,'Sex: Total','Age (Youth, adults): 15-24')

    info2 = select_info(country,'Sex: Total','Age (Youth, adults): 25+')
   
    if country =='Korea, Republic of': 
        country = 'South Korea'

        
    fig, ax = plt.subplots()
    
    plt.xticks(rotation=90)

    ax.plot(info1['time'],info1['obs_value'], label='Age: 15-24') 
    ax.plot(info2['time'],info2['obs_value'], label='Age: +25') 
    plt.ylabel('Unemployment')
    plt.title('Unemployment ' + country)


plt.legend()


```

```python
# Estimated Inflation

inflation = pd.read_csv('#task7_wage_loss_data_source/IndividualDescriptorData/inflation-rate.csv')

inflation
```

```python
# Selecting countries

inflation = inflation[(inflation['Country']=='Australia') | (inflation['Country']=='Denmark')
            | (inflation['Country']=='India')
            | (inflation['Country']=='Japan')
            | (inflation['Country']=='South Korea')
            | (inflation['Country']=='Sweden')
            | (inflation['Country']=='United States')]

inflation
```

```python
# Selecting inflation in Q2/20, Q3/20, Q4/20, Q1/21
#iloc[rows,columns]

inflation.iloc[:,3:]
```

```python
# Plot inflation

fig, ax = plt.subplots(1, 1, figsize=(10,6)) # make the figure with the size 10 x 6 inches

for index, row in inflation.iterrows():

    yaxis=row.iloc[3:]
    
#    plt.xticks(rotation=90)
    
    xaxis = ['Q2/20', 'Q3/20', 'Q4/20', 'Q1/21']
    
    ax.plot(xaxis, yaxis, label=row['Country']) 
    
plt.title('Inflation rate Estimation')

plt.ylabel('Inflation rate')


fig.legend(loc='center right', borderaxespad=0.2)

plt.subplots_adjust(right=0.85)


```
