```python
import pandas as pd
sdata = pd.read_csv("final-clean-sentiment_class.csv")
sdata.drop('Unnamed: 0', axis=1, inplace=True)
sdata.clean_tweet4 = sdata.clean_tweet4.astype(str)
sdata.head(2)
```

```python
# List from youtube group
mental_health_keywords = ['depression', 'mental', 'illness', 'unalive', 'social', 'anxiety', 'loneliness', 'stress', 'lonely', 'isolation', 'suicide', 
                            'abuse', 'death', 'post', 'traumatic', 'stress', 'disorder', 'no', 'motivation', 'therapy', 'trauma', 'counselling', 'mood', 
                            'disorder', 'mood', 'swings', 'mental', 'health', 'angst', 'emotion', 'phobia', 'addiction', 'stigma', 'self-harm', 'neurosis', 
                            'abuse', 'disorder', 'dependence', 'socialize', 'help', 'dead', 'melancholia', 'dysthemia', 'tired', 'trapped', 'paranoia', 
                            'overwhelmed', 'irritable', 'bipolar', 'psychologist', 'well-being', 'imh', 'sos', 'counsellor', 'toxic']
def check_mh(text):
    return any(word in text for word in mental_health_keywords)

covid_list = ['covid', 'coronavirus', 'covid-19', 'pandemic']
def check_cl(text):
    return any(word in text for word in covid_list)
    
sg_list = ['singapore', 'singaporean','sg']
def check_sg(text):
    return any(word in text for word in sg_list)


```

```python
sdata['check_mental_health'] = sdata.clean_tweet4.apply(lambda x: check_mh(x))
sdata['check_covid'] = sdata.clean_tweet4.apply(lambda x: check_cl(x))
sdata['check_sg'] = sdata.clean_tweet4.apply(lambda x: check_sg(x))
```

```python
relevant_list = [sdata.check_mental_health[i] and sdata.check_covid[i] and sdata.check_sg[i] for i in range(len(sdata))]
sum(relevant_list)
```

```python
sdata["relevance"] = relevant_list
```
