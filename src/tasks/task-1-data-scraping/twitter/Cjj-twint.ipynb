```python
kw= pandas.read_csv("./Output/Keyword-list.csv")
```

```python
Mental_Health_Keywords = kw["Mental Health Keywords"][:-2].tolist()
Pandemic_Keywords = kw["Pandemic Keywords"].tolist()
Mental_Health_Keywords
Pandemic_Keywords
```

```python
# Looping through Pandemic_Keywords list, combine all into a file 

for word in Pandemic_Keywords:
    c = twint.Config()
    c.Search = word #search keyword
    c.Since = '2021-01-01'
    c.Until="2021-12-31"
    c.Geo="1.290270,103.851959,20km"
    c.Lang="en"
    c.Filter_retweets = True 
    c.Hide_output = True
    c.Count = True
    c.Pandas = True
    c.Stats = True
    c.Limit = 3000
    twint.run.Search(c)
    df = twint.storage.panda.Tweets_df
    df.to_csv('pk-keywords-21.csv', mode='a', header=False)
    
```
