```python
import pandas as pd
# Check the above link for other installation instructions.
```

```python
mh20=pd.read_csv("/content/mh-keywords-20.csv",encoding= ('latin-1'))
mh20.head()
```

```python
mh20.columns
```

```python
#Remove irrelevant features
columns_keep = [ 'date', 'tweet', 'language', 'hashtags', 'username', 'link',
                'nlikes', 'nreplies', 'nretweets', 'search']

mh20_1= mh20[columns_keep]
```

```python
# mh20_2 Keep only eng tweets 
mh20_2 = mh20_1[mh20_1.language == 'en' ]
mh20_2.language.value_counts()
mh20_2.shape
```

```python
mh20_2.head()
```

```python
# mh20_3 Remove rows that contain certain words  

words = ['Johor', 'Malaysia']
mh20_3 = mh20_2[~mh20_2.tweet.str.contains('|'.join(words))]
mh20_3.shape
mh20_3.to_csv("mh20_3.csv")
```

```python
mh20_3["tweet"].head(20)
```

```python
# dropping ALL duplicate tweets
mh20_3.drop_duplicates(subset ="tweet",
                     keep = False, inplace = True)
```

```python
#list of keyworad used in search
list_of_keywoards=mh20_3["search"].unique()
```

```python
#replace multiple ????? with apostrophe
mh20_3['tweet'] = mh20_3['tweet'].str.replace(r"[?????]", "'",regex=True)
mh20_3['tweet'] = mh20_3['tweet'].str.replace(r"['']", '')
```

```python
mh20_3.shape
```

```python
#drop ads that appear as tweets
word= 'http'
mh20_3 = mh20_3[~mh20_3["tweet"].str.contains(word)]
mh20_3.shape
```

```python
mh20_3.to_csv("mh20_cleaned.csv")
```
