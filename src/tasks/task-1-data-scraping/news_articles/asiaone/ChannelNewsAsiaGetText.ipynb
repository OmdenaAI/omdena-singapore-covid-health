```python
import csv
import json
import os.path
import pandas as pd
import random

from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from time import sleep
```

```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager

driver = webdriver.Chrome(ChromeDriverManager().install())
```

```python
data = pd.read_csv('mental-health-urls.csv', index_col=0)
```

```python
def get_details(url, title):
    result = dict()
    target_page = url
    driver.get(target_page)
    sleep(random.randint(12, 18))
    
    categories = []
    website_categories = driver.find_elements_by_class_name("link-list__link")
    
    for _website_categories in website_categories:  
        link = _website_categories.get_attribute('href')
        if _website_categories.text.strip() != "" and link.startswith("https://www.channelnewsasia.com"):
            categories.append(_website_categories.text)
        
    article_category = driver.find_elements_by_class_name("article__category")
    
    tag = ''
    for _article_category in article_category:  
        tag = _article_category.text
        break
    
    time = driver.find_elements_by_css_selector('time')
    
    ## title, url, date, text, covid (TRUE/FALSE values) 
    
    elem = driver.find_elements_by_css_selector('p')
    body = ''
    for el in elem:
        if el.text == 'Where to get help:':
            break
        body += el.text
    
    
    covid = False
    
    if 'COVID-19' in categories:
        covid = True
    
    result = {
        'title': title,
        'url': url, 
        'date': time[0].text,
        'text': body,
        'covid': covid,
        'tagged_topics': categories,
        'sub_category': tag,
    }
    
    return result
```

```python
overall_result = []
counter = 0
for key, value in data.iterrows():
    counter += 1
    print(counter)
    result = get_details(value['url'], value['title'])
    overall_result.append(result)

pd.DataFrame(overall_result)
```

```python
overall_result_df = pd.DataFrame(overall_result)
```
